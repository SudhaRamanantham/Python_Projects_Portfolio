# Clean the Polling data

## Read the data

```python
import pandas as pd
```

```python
polls = pd.read_csv('president_polls_2016.csv')
polls.head()
```

## Examine the data

```python
# run the info() method on the data

polls.info()
```

```python
# run the nunique() method on the data

polls.nunique()
```

## Drop columns and rows

```python
# drop all rows except where the column is "now-cast"

polls = polls.query("type == 'now-cast'")
polls.head()
```

```python
# run the nunique() method again

polls.nunique()
```

```python
# drop columns with a single value

polls = polls.drop(columns = ['cycle','branch','type','matchup',\
                          'forecastdate','multiversions','timestamp'])
polls.head()
```

```python
# drop rows with "U.S." for the state

polls = polls.query("state != 'U.S.'")
polls.head()
```

## Rename columns

```python
# rename all the rawpoll_* columns

polls = polls.rename(columns = {'rawpoll_clinton':'clinton_pct','rawpoll_trump':'trump_pct',\
                            'rawpoll_johnson':'johnson_pct','rawpoll_mcmullin':'mcmullin_pct'})
polls.head()
```

## Fix data types

```python
# run the info() method again

polls.info()
```

```python
# create a list of columns that should be datetimes

dateTimesCol = ['startdate','enddate','createddate']
print(f'A list of columns that should be datetimes: \n{dateTimesCol}')
```

```python
# convert the columns to datetime

polls[dateTimesCol].apply(pd.to_datetime)
```

```python
# convert the state and population columns to categories

colConvert = {'state':'category','population':'category'}

polls = polls.astype(colConvert)
polls.info()
```
