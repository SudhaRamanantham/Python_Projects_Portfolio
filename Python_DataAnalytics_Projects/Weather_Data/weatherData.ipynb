```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
weatherData = pd.read_csv(r'C:\Users\sudha\Documents\Python\UCSD_Course\DataAnalyticsUsingPython\Assignments\Sudha_spring2023_Qz3\seattle_WA.csv')
weatherData
```

```python
# Adding new column 'daysOfWeek' to weatherData

weatherData['datetime'] = pd.to_datetime(weatherData['datetime'])
weatherData['daysOfWeek'] = weatherData['datetime'].dt.day_name()
weatherData
```

```python
# Shifting column 'daysOfWeek' to second position

daysColumn = weatherData.pop('daysOfWeek')
weatherData.insert(2,'daysOfWeek',daysColumn)
print("After Shifting column to second position")
weatherData
```

```python
# Creating new data frame with 'daysOfWeek' column 

newWeatherData=pd.DataFrame(weatherData)
newWeatherData
```

```python
# After creating the new data frame save that to New_Seattle_WA.csv file

newWeatherData.to_csv('New_Seattle_WA.csv',header=True)
newWeatherData
```

```python
'''Plot two graphs such as minimum and maximum temperature within the same plot showing with 
two different colors.'''

# Scatter Plot

fig = plt.figure(figsize=(10,6))

plt.scatter(newWeatherData['daysOfWeek'],newWeatherData['tempmin'],c='m',label='Minimum Temperature')
plt.scatter(newWeatherData['daysOfWeek'],newWeatherData['tempmax'],c='g',label='Maximum Temperature')

plt.title('Weather Forecast for Seatle, WA')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
# Line Plot

fig = plt.figure(figsize=(12,6))

plt.plot(newWeatherData['daysOfWeek'],newWeatherData['tempmin'],label='Minimum Temperature',color='y')
plt.plot(newWeatherData['daysOfWeek'],newWeatherData['tempmax'],label='Maximum Temperature',color='r')

plt.title('Weather Forecast for Seatle, WA')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```
